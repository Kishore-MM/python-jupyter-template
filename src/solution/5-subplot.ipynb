### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 50)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
x_scatter = np.random.rand(10)
y_scatter = np.random.rand(10)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart in the first subplot
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Chart')

# Bar chart in the second subplot
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Chart')

# Scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')

# Pie chart in the fourth subplot
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Show the plot
plt.show()
```
